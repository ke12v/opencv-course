```python
import os
from pathlib import Path
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Base directory: expects you to have this repo uploaded/checked out in Colab
BASE_DIR = Path('.')
IMG_PATH = BASE_DIR / 'Resources' / 'Photos' / 'park.jpg'
OUT_DIR = BASE_DIR / 'Section 1 output'
OUT_DIR.mkdir(parents=True, exist_ok=True)

if not IMG_PATH.exists():
    raise FileNotFoundError(f'Could not find image: {IMG_PATH.resolve()}')

def show_bgr(title, img_bgr):
    img_rgb = cv.cvtColor(img_bgr, cv.COLOR_BGR2RGB)
    plt.figure(figsize=(6, 6))
    plt.title(title)
    plt.axis('off')
    plt.imshow(img_rgb)
    plt.show()

def show_gray(title, img_gray):
    plt.figure(figsize=(6, 6))
    plt.title(title)
    plt.axis('off')
    plt.imshow(img_gray, cmap='gray')
    plt.show()

img = cv.imread(str(IMG_PATH))
show_bgr('Park', img)
```

```python
# Grayscale
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
show_gray('Gray', gray)
cv.imwrite(str(OUT_DIR / 'basic_functions_Gray.png'), gray)
```

```python
# Blur
blur = cv.GaussianBlur(img, (7, 7), cv.BORDER_DEFAULT)
show_bgr('Blur', blur)
cv.imwrite(str(OUT_DIR / 'basic_functions_Blur.png'), blur)
```

```python
# Canny edges
canny = cv.Canny(blur, 125, 175)
show_gray('Canny Edges', canny)
cv.imwrite(str(OUT_DIR / 'basic_functions_Canny_Edges.png'), canny)
```

```python
# Dilate
dilated = cv.dilate(canny, (7, 7), iterations=3)
show_gray('Dilated', dilated)
cv.imwrite(str(OUT_DIR / 'basic_functions_Dilated.png'), dilated)
```

```python
# Erode
eroded = cv.erode(dilated, (7, 7), iterations=3)
show_gray('Eroded', eroded)
cv.imwrite(str(OUT_DIR / 'basic_functions_Eroded.png'), eroded)
```

```python
# Resize
resized = cv.resize(img, (500, 500), interpolation=cv.INTER_CUBIC)
show_bgr('Resized', resized)
cv.imwrite(str(OUT_DIR / 'basic_functions_Resized.png'), resized)
```

```python
# Cropping
cropped = img[50:200, 200:400]
show_bgr('Cropped', cropped)
cv.imwrite(str(OUT_DIR / 'basic_functions_Cropped.png'), cropped)
```
